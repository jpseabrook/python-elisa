```python
import pandas as pan
import csv
```

```python
file = 'files/input.txt'
#'A','B','C','D','E','F','G','H'
#open(file, 'r', encoding='utf-16').readlines()
```

```python
# read the tsv into a reader, excluded columns/rows i didnt need, 
reader = pan.read_csv(file, encoding='utf-16', sep='\t', header=None, 
                      error_bad_lines=False, skiprows=2, usecols=list(range(2,14)))
# drop empty rows
reader = reader.dropna()
# drop duplicate column number rows, don't keep first occurance
reader = reader.drop_duplicates(keep=False)
# an attempt at labeling all the rows with their letter
dups_label = range(1,41)
index = pan.Index(
    ['A','B','C','D','E','F','G','H'] * int((len(reader)/8))
)
# applying the row labels
reader = reader.set_index(index)
```

```python
# put that into a csv
reader.to_csv('files/output.txt', header=list(range(1,13)))
cleaned = pan.read_csv('files/output.txt', index_col=0)
```
